```python
total_documents = pd.read_csv('/home/nazma/NLP-based-Email-Content-Digital-Forensic-using-ML-model-and-Explainable-AI/merged_topic_features_labeled.csv')
total_documents
```

```python
# For convinience, divide the dataframe cc based on two labels. 

nor_obs = total_documents.loc[total_documents.labeled==False]    #Data frame with normal observation
ano_obs = total_documents.loc[total_documents.labeled==True]    #Data frame with anomalous observation

```

```python


# Once class SVM is trained with the observations of only one class. In this case, the algorithm is trained with 
# first 200,000 observation of normal transactions. The remaining observation is merged with the anomalous observation 
# to create a test set. 

train_feature = nor_obs.loc[0:200000, :]
train_feature = train_feature.drop('labeled', axis=1)
Y_1 = nor_obs.loc[200000:, 'labeled']
Y_2 = ano_obs['labeled']

```

```python
# Creating test observations/features

X_test_1 = nor_obs.loc[200000:, :].drop('labeled',axis=1)
X_test_2 = ano_obs.drop('labeled',axis=1)
# X_test = X_test_1.append(X_test_2)
X_test = pd.concat([X_test_1, X_test_2], ignore_index=True)
```

```python
X_test
```

```python

from sklearn import svm

# Setting the hyperparameters for Once Class SVM

oneclass = svm.OneClassSVM(kernel='linear', gamma=0.001, nu=0.95)

# I have used various combination of hyperparameters like linear, rbf, poly, gamma- 0.001, 0.0001, nu- 0.25, 0.5, 0.75, 0.95
# This combination gave me the most satisfactory results.# The remain data set is (after 200,000 observations) are appended with anomalous observations

Y_1 = nor_obs.loc[200000:, 'labeled']
Y_2 = ano_obs['labeled']

# Y_test= Y_1.append(Y_2)
Y_test = pd.concat([Y_1, Y_2], ignore_index=True)

#Y_test is used to evaluate the model
```

```python
# Training the algorithm with the features. 
# This stage is very time consuming processes.

oneclass.fit(train_feature)
```

```python


# Test the algorithm on the test set

fraud_pred = oneclass.predict(X_test)

```

```python
import numpy as np
# Check the number of outliers predicted by the algorithm

unique, counts = np.unique(fraud_pred, return_counts=True)
print (np.asarray((unique, counts)).T)
```

```python


#Convert Y-test and fraud_pred to dataframe for ease of operation

Y_test= Y_test.to_frame()
Y_test=Y_test.reset_index()
fraud_pred = pd.DataFrame(fraud_pred)
fraud_pred= fraud_pred.rename(columns={0: 'prediction'})

```

```python
##Performance check of the model

TP = FN = FP = TN = 0
for j in range(len(Y_test)):
    if Y_test['labeled'][j]== False and fraud_pred['prediction'][j] == 1:
        TP = TP+1
    elif Y_test['labeled'][j]== False and fraud_pred['prediction'][j] == -1:
        FN = FN+1
    elif Y_test['labeled'][j]== True and fraud_pred['prediction'][j] == 1:
        FP = FP+1
    else:
        TN = TN +1
print (TP,  FN,  FP,  TN)
```

```python


# Performance Matrix

accuracy = (TP+TN)/(TP+FN+FP+TN)
print (accuracy)
precision = TP/(TP + FP)
print(precision)
recall = TP/(TP + FN)
print(recall)
f1_score = (2 * precision * recall) / (precision + recall)
print(f1_score)
sensitivity = TP/(TP+FN)
print (sensitivity)
specificity = TN/(TN+FP)
print (specificity)

```
