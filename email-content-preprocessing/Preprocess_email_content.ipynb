```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

def remove_emails(text):
    # Regular expression pattern to match email addresses
    pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
    # Find the matched email addresses
    # matches = re.findall(pattern, text)
    # print the matched email for check
    # print(matches)
    # Replace email addresses with an empty string
    return re.sub(pattern, '', text)

def remove_stopwords(content):
    "Return the content after remvoing stopwords"
    content_tokens = word_tokenize(content)
    filtered_content = [word for word in content_tokens if not word in all_stopwords]
    return " ".join(filtered_content)

#remove emails
Basic = Basic.apply(remove_emails)

#removing stopwords
Basic = Basic.apply(remove_stopwords)

#removing Punctuations
Basic = Basic.apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '', x))

#removing digits
Basic = Basic.apply(lambda x: re.sub('\w*\d\w*','', x))
```

```python
def split_chunks(data, num_chunks):
  """Splits a list (or any iterable) into chunks of a specified size."""
  chunk_size = len(data) // num_chunks  # Integer division for equal chunks (rounded down)
  return [data[i:i + chunk_size] for i in range(0, len(data), chunk_size)]

num_chunks = 10

chunks = split_chunks(basic, num_chunks)
print(chunks)
```

```python
def lemmatize_text(text):

    if isinstance(text, str):
        doc = nlp(text)
        lemmatized_text = " ".join([token.lemma_ for token in doc])
        return lemmatized_text
    else:
        return ""  # Handle non-string elements (e.g., return empty string)

"""# Split train_data into chunks
chunk_size = int(len(basic) / 10)  # Calculate chunk size for 10 chunks
chunks = [basic[i:i + chunk_size] for i in range(0, len(basic), chunk_size)]
"""


# Lemmatize chunks and update original DataFrame
for i, chunk in enumerate(chunks):
  lemmatized_chunk = chunk["content"].apply(lemmatize_text)
  basic.loc[chunk.index, "content"] = lemmatized_chunk
    


# # Lemmatize each chunk
# for i, chunk in enumerate(chunks):
# #   chunk['lemmatized_content'] = chunk['content'].apply(lemmatize_text)
#   chunk.loc[:, 'lemmatized_content'] = chunk['content'].apply(lemmatize_text)
# #   chunk['lemmatized_content'] = chunk['lemmatized_content'].set_index(range(i*chunk_size, (i+1)*chunk_size))
#   # Update the corresponding chunk in train_data using .loc (fixing the warning)
#   X_train.loc[i*chunk_size: (i+1)*chunk_size-1, 'lemmatized_content'] = chunk['lemmatized_content']

# # Alternatively, create a new DataFrame with all chunks (Option 2)
# # new_train_data = pd.concat(chunks, ignore_index=True)

# X_train  # Print the modified train_data (with "lemmatized_content" column)

```
