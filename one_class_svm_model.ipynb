```python
import pandas as pd
from sklearn.preprocessing import StandardScaler  # Optional
from sklearn.svm import OneClassSVM
```

```python
# Assuming "normal_documents" ids DataFrame with topic features
normal_documents = pd.read_csv('/content/drive/MyDrive/topic_features_00.csv')
normal_documents
```

```python
normal_features = normal_documents.drop("doc_id", axis=1)  # removing the document id

normal_features
```

```python
# Scale features (optional)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(normal_features)  # Or use normal_features directly

# Train one-class SVM model
model = OneClassSVM(nu=0.1)  # Adjust nu based on the proportion of anomalies you expect
model.fit(scaled_features)
```

after topic features of test data is calculated

```python
new_documents = pd.read_csv('/content/drive/MyDrive/test_topic_features_final_labeled.csv')
new_documents
```

```python
#  "new_documents" is  DataFrame with topic features for new documents
new_features = new_documents.drop(columns=["doc_id","labeled", "Unnamed: 0"] )

new_features
```

```python
# Scale new features
new_scaled_features = scaler.transform(new_features)

# Predict labels for new documents (1 for inliers, -1 for outliers)
new_document_labels = model.predict(new_scaled_features)

# Convert predicted labels to Pandas Series
new_document_labels = pd.Series(new_document_labels)

# Identify documents classified as outliers (label -1)
outlier_indices = new_document_labels[new_document_labels == -1].index

outlier_documents = new_documents.iloc[outlier_indices]  # Assuming row indices for documents

print("Potential anomaly documents:")
print(outlier_documents)
```

```python
from sklearn.preprocessing import LabelEncoder

true_labels = new_documents["labeled"]  # Extract true labels


# Encode true labels
le = LabelEncoder()
encoded_labels = le.fit_transform(true_labels)


predicted_labels = new_document_labels

```

```python
# Calculate evaluation metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(encoded_labels, predicted_labels)
precision = precision_score(encoded_labels, predicted_labels, average='macro')
recall = recall_score(encoded_labels, predicted_labels, average='macro')
f1 = f1_score(encoded_labels, predicted_labels, average='macro')

# Print evaluation results
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
